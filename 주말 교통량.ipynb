# 광진구 주말 대중교통량

```python
import pandas as pd
import numpy as np
```

```python
traffic = pd.read_csv("서울시 교통정보.csv")
traffic.head()
```

## 광진구의 시간대별 합계 교통량 

```python
T=(
    traffic
    .groupby(['자치구명'])
    .agg('mean')
    .loc['광진구','합계00H':'합계23H']
    .round(0)
    .astype(int)

)
T
```

```python
import matplotlib.pyplot as plt
 
# 한글 폰트 사용을 위해서 세팅
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

fig, ax = plt.subplots(figsize=(20, 10))
plt.bar(T.index, T)
```

## 서울시 교통정보 데이터 중 광진구 데이터만 가져오기

```python
data = pd.read_csv("서울시 교통정보.csv",index_col='자치구명')
M=data.loc['광진구',:]
M
```

## 행정동별로 합계량 구하기

```python
M=(
    M
    .groupby(['행정동명'])
    .agg('mean')
    .loc[:,:'합계']
    .round(0)
    .astype('int')
    
)
M
```

```python
import matplotlib.pyplot as plt
 
# 한글 폰트 사용을 위해서 세팅
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

fig, ax = plt.subplots(figsize=(12, 10))
plt.bar(M.index,M.합계)
```

# 광진구 월요일 대중교통량

```python
m_traffic = pd.read_csv('월요일 서울시 교통량.csv')
m_traffic.head()
```

## 월요일 데이터 가져오기

```python
m_traffic = m_traffic[m_traffic['기준일'] == 20230529]
m_traffic.head()
```

## 광진구 데이터만 가져오기

```python
m_traffic=m_traffic[m_traffic['자치구명'] == '광진구']
m_traffic.head()
```

## 광진구의 시간대별 합계 교통량

```python
T=(
    m_traffic
    .groupby(['자치구명'])
    .agg('mean')
    .loc['광진구','합계00H':'합계23H']
    .round(0)
    .astype(int)

)
T
```

```python
import matplotlib.pyplot as plt
 
# 한글 폰트 사용을 위해서 세팅
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

fig, ax = plt.subplots(figsize=(20, 10))
plt.bar(T.index, T)
```

## 행정동별 합계량 구하기

```python
m_traffic=m_traffic.loc[:,['행정동명','합계']]
m_traffic.sort_values(by = ['행정동명','합계'], ascending = True, inplace = True)
m_traffic
```

```python
import matplotlib.pyplot as plt
 
# 한글 폰트 사용을 위해서 세팅
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

fig, ax = plt.subplots(figsize=(12, 10))
plt.bar(m_traffic.행정동명,m_traffic.합계)
```
